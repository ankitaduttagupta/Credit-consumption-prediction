```python
def rmsle(y_test, y_pred):
    
    sum=0.0
    
    for x,y in zip(y_test, y_pred):
        
        if x < 0 or y < 0: #check for negative values. 
            continue
        
        p = np.log(y+1)
        r = np.log(x+1)
        sum += (p - r)**2
    
    return (sum / len(y_pred)) ** 0.5
```
